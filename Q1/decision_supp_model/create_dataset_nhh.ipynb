```python
import pandas as pd
import numpy as np
import re
import unicodedata
```

```python
lka = lka.copy()
eth = eth.copy()
nga = nga.copy()
zwe = zwe.copy()
moz = moz.copy()
yem = yem.copy()

lka["country"] = "Sri Lanka"
eth["country"] = "Ethiopia"
nga["country"] = "Nigeria"
zwe["country"] = "Zimbabwe"
moz["country"] = "Mozambique"
yem["country"] = "Yemen"

pop_df = pd.concat([lka, eth, nga, zwe, moz, yem], ignore_index=True)
pop_df = pop_df[["country", "name", "totale"]].copy()

pop_df.head()
```

```python

experiment_to_country = {
    "eth_micron": "Ethiopia",
    "nga_micron": "Nigeria",
    "zwe_mics": "Zimbabwe",
    "moz_vam": "Mozambique",
    "yem_mvam": "Yemen",
    "lka_micron": "Sri Lanka",
    "lka_vam": "Sri Lanka",
    "nga_mics": "Nigeria",
}

final_df = pd.read_csv('/data/shared/fsibilla/clean_code/Q1/decision_supp_model/adm1_level_summary.csv')

final_df.head()
```

```python
import pandas as pd
import unicodedata

def normalize_name(s):
    if pd.isna(s):
        return s
    s = str(s).strip().lower()
    s = unicodedata.normalize("NFKD", s).encode("ascii", "ignore").decode("utf-8")
    s = " ".join(s.split())
    return s

# manual corrections for df_final adm1 values
adm1_corrections = {
    ("eth", "B. Gumuz"): "Benishangul Gumz",
    ("eth", "SNNPR"): "SNNP",
    
    ("moz", "Maputo Cidade"): "Maputo City",
    ("moz", "Maputo Provínci"): "Maputo",
    
    ("nga", "Abuja"): "Federal Capital Territory",
    ("nga", "Nassarawa"): "Nasarawa",
    
    # Yemen: these target names must match the names in your yem dataframe
    ("yem", "Abyan"): "Abyan",
    ("yem", "Aden"): "Aden",
    ("yem", "Al Hodeidah"): "Al Hodeidah",
    ("yem", "Al Jawf"): "Al Jawf",
    ("yem", "Al Maharah"): "Al Maharah",
    ("yem", "Dhamar"): "Dhamar",
    ("yem", "Hadramawt"): "Hadramawt",
    ("yem", "Lahj"): "Lahj",
    ("yem", "Raymah"): "Raymah",
    ("yem", "Sana'a"): "Sana'a",
    ("yem", "Shabwah"): "Shabwah",
    ("yem", "Socotra"): "Socotra",
    ("yem", "Ta'iz"): "Ta'iz",}


df_final = final_df.copy()
df_final["country"] = df_final["experiment"].str.split("_").str[0]

# create corrected adm1
df_final["adm1_corrected"] = df_final.apply(
    lambda row: adm1_corrections.get((row["country"], row["adm1"]), row["adm1"]),
    axis=1
)

df_final["adm1_norm"] = df_final["adm1_corrected"].apply(normalize_name)
```

```python
dfs = {
    "eth": eth.copy(),
    "nga": nga.copy(),
    "zwe": zwe.copy(),
    "lka": lka.copy(),
    "moz": moz.copy(),
    "yem": yem.copy()
}

for code, d in dfs.items():
    d["country"] = code
    d["name_norm"] = d["name"].apply(normalize_name)

lookup = pd.concat(dfs.values(), ignore_index=True)[["country", "name", "name_norm", "totale"]]

merged = df_final.merge(
    lookup,
    how="left",
    left_on=["country", "adm1_norm"],
    right_on=["country", "name_norm"]
)

merged = merged.drop(columns=["name_norm"])
```

```python
unmatched = merged[merged["totale"].isna()][["experiment", "adm1", "country"]].drop_duplicates()
print(unmatched)
```
